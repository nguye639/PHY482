```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def emf(w,b,v,a,x,t):
    e = -w*(b*v+a*x+2*a*v*t)
    return e

w = 2
b = .02
v = 2
x = 2
a = 2
t = np.linspace(0,10,100)

for i in range(0,10,1):
    plt.plot(t,emf(w,b,v,a**i,x,t))

```

```python
for i in range(0,10,1):
    plt.plot(t,emf(w,b,v**i,a,x,t))

```

For both alpha and velocity, I found that as the order of magnitude increases the emf tends to zero.
